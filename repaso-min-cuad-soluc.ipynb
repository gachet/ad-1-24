# <span style="color:red">Ajuste por mínimos cuadrados MONOVARIABLE considerando una recta de la forma </span>
$$F(x)= \alpha x+\beta$$
### $$Utilizando\hspace{0.5cm} ecuaciones\hspace{0.5cm} NORMALES$$

#### Considerando po ejemplo  tres pares de puntos $(x_{1},y_{1}),(x_{2},y_{2}),(x_{3},y_{3})$ el conjunto de todas las líneas que pasan por un par de puntos sería
$$
\begin{bmatrix}
x_{1}& 1 \\
x_{2}& 1  \\
x_{3}& 1
\end{bmatrix}
\begin{bmatrix}
\alpha\\
\beta
\end{bmatrix} =\begin{bmatrix}
y_{1}\\
y_{2}\\
y_{3}
\end{bmatrix}
$$

##Se puede generalizar a  m  pares de datos
$$
\begin{bmatrix}
x_{1}& 1 \\
x_{2}& 1  \\
x_{3}& 1 \\
\cdot & \cdot \\
\cdot & \cdot \\
x_{m} & 1
\end{bmatrix}
\begin{bmatrix}
\alpha\\
\beta
\end{bmatrix} =\begin{bmatrix}
y_{1}\\
y_{2}\\
y_{3} \\
\cdot \\
\cdot \\
y_{m}
\end{bmatrix}
$$

El sistema matricial se puede escribir como $Ac = y$ que no se puede resolver por Gauss por ser sobredeterminado. Aplicando el criterio de mínimos cuadrados llegamos a la ecuación matricial:

$$\left ( A^{T}A \right )c=A^{T}y$$

##El vector solución &nbsp;$c$&nbsp; se calcula como &nbsp;&nbsp;&nbsp; $c=\left ( A^{T}A \right )^{inv} A^{T}y$

# <span style="color:red">Ajuste por mínimos cuadrados mediante funciones base (Ajuste Polinómico)
### Utilizando ecuaciones NORMALES


##Considerar una función de ajuste con tres funciones base
$$
y = F(x) = c1f1(x) + c2f2(x) + c3f3(x).
$$
Asumimos que  F(x) actua como un interpolante. Entonces, el sistema

$$
c1f1(x1) + c2f2(x1) + c3f3(x1) = y1,
c1f1(x2) + c2f2(x2) + c3f3(x2) = y2,
...
c1f1(xm) + c2f2(xm) + c3f3(xm) = ym.
$$

se satisface.

##<font color='blue'>Igual que en el caso de la recta  tenenos mas datos que ecuaciones, éste es un sistema sobredeterninado y la función de ajuste F(x) no pasa por todos los puntos $y_{i}$. El sistema se puede escribir como</font>

$$
Ac = y
$$

##Se puede generalizar a $m$ pares de datos

$$
\begin{bmatrix}
f_{1}(x_{1})& f_{2}(x_{1})&\cdot\cdot\cdot &f_{n}(x_{1})   \\
f_{1}(x_{2})& f_{2}(x_{2})&\cdot\cdot\cdot &f_{n}(x_{2})  \\
f_{1}(x_{3})& f_{2}(x_{3})&\cdot\cdot\cdot &f_{n}(x_{3}) \\
\cdot & \cdot \\
\cdot & \cdot \\
f_{1}(x_{m})& f_{2}(x_{m})&\cdot\cdot\cdot &f_{n}(x_{m})
\end{bmatrix}
\begin{bmatrix}
c_{1}\\
c_{2} \\
\cdot \\
\cdot \\
c_{n}
\end{bmatrix} =\begin{bmatrix}
y_{1}\\
y_{2}\\
y_{3} \\
\cdot \\
\cdot \\
y_{m}
\end{bmatrix}
$$

Igual que en el caso de la recta, el sistema matricial se puede escribir como $Ac = y$ que no se puede resolver por Gauss por ser sobredeterminado. Aplicando el criterio de mínimos cuadrados llegamos a la ecuación matricial

$$\left ( A^{T}A \right )c=A^{T}y$$

##El vector solución &nbsp;$c$&nbsp; se calcula como &nbsp;&nbsp;&nbsp; $c=\left ( A^{T}A \right )^{inv} A^{T}y$

## Ejemplo: Sea la función $ y = c_{1}x^{2}+c_2{x}+c_{3}$

###  Tenemos las funciones base  $f_{1}(x) = x^{2},\:\:f_{2}(x) = x\:\:y\:\:f_{3}(x)=1$

##La matriz A estaría formada por

$$
\begin{bmatrix}
x_{1}^{2}& x_{1}& 1\\
x_{2}^{2}& x_{2}& 1\\
\cdot & \cdot & \cdot\\
\cdot & \cdot &\cdot\\
x_{m}^2 & x_{m}& 1
\end{bmatrix}
$$

## Validez del Ajuste: Utilizamos El coeficiente de determinación se puede calcular  como:
$$
R^{2}=1- \frac{\parallel r \parallel_{2}^{2}}{\displaystyle\sum_{i=1}^{m}\left(y_{i}-\bar{y}\right)^{2}}
$$

##Siendo r el vector de residuos, m el número de datos y n el número de funciones base, Mientras más se aproxime al valor 1 el ajuste será mejor.

```python
input_data=np.loadtxt("/content/drive/MyDrive/minimos/productos.dat",skiprows=1)
x=input_data[:,0]
y=input_data[:,1]
x = x.reshape(-1,1)
y = y.reshape(-1,1)
print('data read in with ',len(x),' points \n')
print('Años Mercado \t Precio')
for i in range(0,len(x)):
    print('%4.1f\t\t'%(x[i]),'%4.3f'%(y[i]))
```

##Paso 2 Formamos nuestra Matriz A  (tener en cuenta que la última columna debe stener  todo los elementos a 1.

```python
unos = np.ones(len(x))
```

```python
A = np.column_stack((x**2,x,unos))
print('Matriz A\n',A)
```

```python
A.shape[1]
```

```python
c = np.dot(np.linalg.inv(np.dot(A.T,A)),np.dot(A.T,y))
print('Vector  c \n',c)
```

### Ecuacion es &nbsp;&nbsp;&nbsp;&nbsp; $$precio = 0.01659años^{2}-0.3458años+1.7$$

```python
c1 = c[0]
c2 = c[1]
c3 = c[2]

y_calc = c1*x**2+c2*x+c3
suma_errores = np.sum((y-y_calc)**2)
suma_errores
```

### El coeficiente de determinación se puede calcular también como:
$$
R^{2}=1-\left( \frac{m-1}{m-n-1} \right) \frac{\parallel r \parallel_{2}^{2}}{\displaystyle\sum_{i=1}^{m}\left(y_{i}-\bar{y}\right)^{2}}
$$

```python
r = y - np.dot(A,c)
m,n = A.shape[0],A.shape[1]

R2=1 - ((m-1)/(m-n-1))*np.linalg.norm(r)**2 / np.sum ((y - np.mean(y))** 2)
print('Coeficiente de determinación es: ',R2)
```
